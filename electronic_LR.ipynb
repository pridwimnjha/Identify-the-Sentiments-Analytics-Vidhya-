```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk.corpus import stopwords
import re
import nltk

# Download necessary resources if not already downloaded
nltk.download('wordnet', quiet=True)
nltk.download('stopwords', quiet=True)


lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def lemmatization(content):
    lemmatized_content = re.sub('[^a-zA-Z]',' ', content)
    lemmatized_content = lemmatized_content.lower()
    lemmatized_content = lemmatized_content.split()
    lemmatized_content = [lemmatizer.lemmatize(word) for word in lemmatized_content if not word in stop_words]
    lemmatized_content = ' '.join(lemmatized_content)
    return lemmatized_content

twitter_data['tweet_lemmatized'] = twitter_data['tweet'].apply(lemmatization)
display(twitter_data[['tweet', 'tweet_lemmatized']].head())
```

```python
X = twitter_data['tweet_lemmatized'].values
Y = twitter_data['label'].values
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split( X , Y , test_size=0.2 , stratify=Y , random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score

#accuracy score in training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy score on the training data :', training_data_accuracy )
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy score on the test data :', test_data_accuracy )
```
